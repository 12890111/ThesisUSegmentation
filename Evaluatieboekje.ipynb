```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, Image, clear_output

df = pd.read_excel('model performances.xlsx')

dropdown = widgets.Dropdown(
    options=df.index.tolist(),
    description='Select Entry:',
    disabled=False,
)

def on_dropdown_change(change):
    clear_output(wait=True)
    display(df)
    display(dropdown)
    
    entry_number = change['new']
    
    img1 = Image(filename=f'run{entry_number}_1.png') 
    img2 = Image(filename=f'run{entry_number}_ROC.png') 
    
    display(img1,img2)

    eval1 = df.loc[entry_number, 'AVG Dice']
    eval2 = df.loc[entry_number, 'Muscle Dice']
    eval3 = df.loc[entry_number, 'Muscle Jacc']
    eval4 = df.loc[entry_number, 'Muscle F1']
    
    print(f"4 label Dice score 1: {eval1}")
    print(f"Muscle Dice score: {eval2}")
    print(f"Muscle IOU: {eval3}")
    print(f"Muscle F1 score: {eval4}")

display(df)
dropdown.observe(on_dropdown_change, names='value')
display(dropdown)
```
